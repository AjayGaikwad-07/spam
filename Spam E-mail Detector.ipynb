```python
with open(r'/content/drive/MyDrive/Document from Jayesh Raygade', 'rb') as f:
    encoding = chardet.detect(f.read())['encoding']
data = pd.read_csv(r'/content/drive/MyDrive/Document from Jayesh Raygade', encoding=encoding)
print(data)


```

```python
data.groupby('category').describe()
```

```python
# turn spam/ham into nummerical data, creating a new column called 'spam'
data['spam'] = data['category'].apply(lambda x: 0 if x == 'spam' else 1)
```

```python
data
```

```python
#Split the data into train/test dataset

x_train, x_test, y_train, y_test = train_test_split(data.massege, data.spam, test_size = 0.25)
```

```python
# Find word count and store data as a matrix
cv = CountVectorizer()
x_train_count = cv.fit_transform(x_train.values)
x_test_count =(x_test.values)
```

```python
#train model
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
#pre-test ham
mail_ham =["hey wanna meet up for the game?"]
mail_ham_count = cv.transform(mail_ham)
model.predict(mail_ham_count)
```

```python
#pre-test spam
mail_spam =["you are win the reward!"]
mail_spam_count =cv.transform(mail_spam)
model.predict(mail_spam_count)
```

```python
# Test the accuracy of the model

x_test_count = cv.transform(x_test)
model.score(x_test_count, y_test)
```
